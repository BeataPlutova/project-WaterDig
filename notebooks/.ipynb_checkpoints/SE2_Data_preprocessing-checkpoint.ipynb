```python
## Part 1: Load raw climate data 
```

```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
# --- Path setup ---
notebook_dir = Path.cwd()                  # /home/jovyan/project-WaterDig/notebooks
project_root = notebook_dir.parent.parent  # /home/jovyan directory

# Raw data file (single CSV)
raw_file = project_root / "raw_data" / "climate_data"/ "jokioinen_daily_1991_2022.csv"

# Output directory
processed_dir = notebook_dir.parent / "processed_data"
processed_dir.mkdir(exist_ok=True)

print("Reading raw file:", raw_file)
print("Processed data will be saved to:", processed_dir)
```

```python
#--- Load raw climate data ---
Climate_data_raw = pd.read_csv(raw_file)
Climate_data_raw["date"] = pd.to_datetime(Climate_data_raw["date"])
Climate_data_raw.head()
```

```python
## Part 2: Pre-processing of raw climate data 
```

```python
# Part 2.1: Column re-naming 
```

```python
rename_map = {
    "rrday": "precipitation(mm)",
    "snow": "snow_depth(cm)",
    "tday": "mean_temperature(°C)",
    "tmin": "min_temperature(°C)",
    "tmax": "max_temperature(°C)"
}

Climate_data_raw = Climate_data_raw.rename(columns=rename_map)
Climate_data_raw.head()
```

```python
# Part 2.2: Handling -1 values in precipitation and snow depth 
```

```python
# By FMI:
# -1  -> 0 mm   (no rain)

if "precipitation(mm)" in Climate_data_raw.columns:
    Climate_data_raw["precipitation(mm)"] = Climate_data_raw["precipitation(mm)"].replace(-1, 0)

if "snow_depth(cm)" in Climate_data_raw.columns:
    Climate_data_raw["snow_depth(cm)"] = Climate_data_raw["snow_depth(cm)"].replace(-1, 0)
```

```python
(Climate_data_raw["precipitation(mm)"] == -1).any()
```

```python
(Climate_data_raw["snow_depth(cm)"] == -1).any()
```

```python
# Part 2.3: Check for NA values in the dataset
```

```python
Climate_data_raw.isna().sum()
```

```python
# By FMI:
# NA  -> observation data is missing - NA values will be kept in the dataset
```

```python
# Part 2.4: Monthly, seasonal, yearly aggregation
```

```python
Climate_data_raw["date"].dtype
```

```python
# Add year column
Climate_data_raw["year"] = Climate_data_raw["date"].dt.year

# Add month column
Climate_data_raw["month"] = Climate_data_raw["date"].dt.month
```

```python
# Monthly climate data
Climate_data_monthly = Climate_data_raw.groupby(
    ["year", "month"]
).agg({
    "precipitation(mm)": "sum",        
    "snow_depth(cm)": "mean",          
    "mean_temperature(°C)": "mean",    
    "min_temperature(°C)": "mean",
    "max_temperature(°C)": "mean"
}).reset_index()

Climate_data_monthly.head()
```

```python
Climate_data_monthly.isna().sum()
```

```python
# Create a season column
Climate_data_raw["season"] = Climate_data_raw["date"].dt.month % 12 // 3 + 1

season_labels = {1: "Winter", 2: "Spring", 3: "Summer", 4: "Autumn"}
Climate_data_raw["season"] = Climate_data_raw["season"].map(season_labels)
```

```python
# Seasonal climate data
Climate_data_seasonal = Climate_data_raw.groupby(
    ["year", "season"]
).agg({
    "precipitation(mm)": "sum",        # total precipitation in season
    "snow_depth(cm)": "mean",          # average seasonal snow depth
    "mean_temperature(°C)": "mean",    # mean seasonal temperature
    "min_temperature(°C)": "mean",
    "max_temperature(°C)": "mean"
}).reset_index()

Climate_data_seasonal.head()
```

```python
# Yearly climate data
Climate_data_yearly = Climate_data_raw.groupby(
    ["year"]
).agg({
    "precipitation(mm)": "sum",        
    "snow_depth(cm)": "mean",          
    "mean_temperature(°C)": "mean",    
    "min_temperature(°C)": "mean",
    "max_temperature(°C)": "mean"
}).reset_index()

Climate_data_yearly.head()
```

```python
# Save processed datasets
Climate_data_monthly.to_csv(processed_dir / "Climate_data_monthly.csv", index=False)
Climate_data_seasonal.to_csv(processed_dir / "Climate_data_seasonal.csv", index=False)
Climate_data_yearly.to_csv(processed_dir / "Climate_data_yearly.csv", index=False)
```
